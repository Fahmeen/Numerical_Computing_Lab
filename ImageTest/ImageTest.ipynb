```python
import numpy as np
import matplotlib.pyplot as plt
def tint(imag, percent):
    tinted_imag = imag + (np.ones(imag.shape) - imag) * percent
    return tinted_imag
image = plt.imread('anime.png')
tinted_image = tint(image, 0.8)
plt.axis("off")
plt.imshow(tinted_image)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def shade(img, percent):
    tinted_img = img * (1 - percent)
    return tinted_img
image = plt.imread('anime.png')
tinted_image = shade(image, 0.7)
plt.imshow(tinted_image)
```
